```python
## create the list of embeddings
embeddings = calculator.calculate_embedding(candidate_questions)
embedding = calculator.calculate_embedding(base_question)[0]
np_embeddings = np.array(embeddings.to('cpu')).astype('float32')
embedding.shape
```

```python
## pick one of the stories as the base for searech
#tokenized_dataset
import faiss
index_flat = faiss.IndexFlatL2(calculator.hidden_size)
res = faiss.StandardGpuResources()
gpu_index_flat = faiss.index_cpu_to_gpu(res, 0, index_flat)
gpu_index_flat.add(np_embeddings)
k = 1
D, I = gpu_index_flat.search(np.expand_dims(embedding.to('cpu'), axis=0), k)
close_question = candidate_questions[I[0][0]]
#print('close text index is ' + str(I[0][1]))
print('close_question qid is: ',candidate_qids[I[0][0]])
print('close_question is ', close_question)
```
